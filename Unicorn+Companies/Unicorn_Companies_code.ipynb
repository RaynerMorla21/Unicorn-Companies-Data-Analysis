```python
def replace_suffixes(funding_str):
    if isinstance(funding_str, str):  # Check if the value is a string
        if 'B' in funding_str:
            return funding_str.replace('B', '000000000')
        elif 'M' in funding_str:
            return funding_str.replace('M', '000000')
    return funding_str  # Return the original value if it's not a string

# Apply the function to the 'Total Funding' and 'Current Valuation' columns
data_df['Funding'] = data_df['Funding'].apply(replace_suffixes)
data_df['Valuation'] = data_df['Valuation'].apply(replace_suffixes)

# Remove any remaining non-numeric characters (e.g., '$')
data_df['Funding'] = data_df['Funding'].replace('[\$,]', '', regex=True)
data_df['Valuation'] = data_df['Valuation'].replace('[\$,]', '', regex=True)

# Convert columns to numeric type
data_df['Funding'] = pd.to_numeric(data_df['Funding'], errors='coerce')  # add errors='coerce' to handle any conversion errors
data_df['Valuation'] = pd.to_numeric(data_df['Valuation'], errors='coerce')

# Calculate ROI and add it as a new column
data_df['ROI'] = (data_df['Valuation'] - data_df['Funding']) / data_df['Funding']

# Rank companies based on their ROI
data_df['ROI Rank'] = data_df['ROI'].rank(ascending=False)

# Sort the DataFrame by ROI to see the ranking clearly
data_df_sorted_by_roi = data_df.sort_values(by='ROI', ascending=False)

# Display the top 10 companies with the highest ROI
print(data_df_sorted_by_roi[['Company', 'Valuation', 'Funding', 'ROI', 'ROI Rank']].head(10))
```

```python


# Convert 'Date Joined' to datetime
data_df['Date Joined'] = pd.to_datetime(data_df['Date Joined'])

# Extract the year from 'Date Joined'
data_df['Year Joined'] = data_df['Date Joined'].dt.year

# Calculate the time to become a unicorn
data_df['Time to Unicorn'] = data_df['Year Joined'] - data_df['Year Founded']

# Use descriptive statistics to find the mean, median, and range
mean_time_to_unicorn = data_df['Time to Unicorn'].mean()
median_time_to_unicorn = data_df['Time to Unicorn'].median()
range_time_to_unicorn = data_df['Time to Unicorn'].max() - data_df['Time to Unicorn'].min()

# Print the descriptive statistics
print(f"Mean time to become a unicorn: {mean_time_to_unicorn:.2f} years")
print(f"Median time to become a unicorn: {median_time_to_unicorn} years")
print(f"Range of time to become a unicorn: {range_time_to_unicorn} years")

# Compare the time-to-unicorn over different periods
# For simplicity, we'll divide the data into two periods: before 2010 and 2010 and after
data_df['Period'] = ['Before 2010' if year < 2010 else '2010 and After' for year in data_df['Year Founded']]

# Calculate the mean time to unicorn for each period
mean_time_by_period = data_df.groupby('Period')['Time to Unicorn'].mean()

# Print the mean time to unicorn by period
print("\nMean time to become a unicorn by period:")
print(mean_time_by_period)
```
